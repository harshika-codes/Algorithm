```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
# -------------------- Synthetic Dataset --------------------

np.random.seed(42)

n_students = 1000

df = pd.DataFrame({

    "Study_Hours": np.random.randint(1,10,n_students),

    "Attendance": np.random.randint(60,100,n_students),

    "Assignments": np.random.randint(1,20,n_students),

    "Projects": np.random.randint(0,10,n_students),

    "Previous_Marks": np.random.randint(40,100,n_students),

    "Internship": np.random.randint(0,3,n_students),

    "Coding_Score": np.random.randint(40,100,n_students)

})

# Target Variable

df["Final_Score"] = (

      df["Study_Hours"]*4

    + df["Attendance"]*0.35

    + df["Assignments"]*1.8

    + df["Projects"]*2

    + df["Previous_Marks"]*0.45

    + df["Internship"]*5

    + df["Coding_Score"]*0.50

    + np.random.normal(0,5,n_students)

)

print(df.head())
```

```python
# -------------------- Features --------------------

X = df.drop("Final_Score",axis=1)

y = df["Final_Score"]
```

```python
# -------------------- Split Dataset --------------------

X_train,X_test,y_train,y_test = train_test_split(

    X,
    y,
    test_size=0.20,
    random_state=42

)
```

```python
# -------------------- Models --------------------

models = {

    "Linear Regression":LinearRegression(),

    "Decision Tree":DecisionTreeRegressor(random_state=42),

    "Random Forest":RandomForestRegressor(
        n_estimators=100,
        random_state=42
    ),

    "KNN":KNeighborsRegressor(),

    "Support Vector Regressor":SVR(),

    "XGBoost":XGBRegressor(
        random_state=42
    )

}
```

```python
# -------------------- Training --------------------

results=[]

for name,model in models.items():

    model.fit(X_train,y_train)

    prediction=model.predict(X_test)

    mae=mean_absolute_error(y_test,prediction)

    mse=mean_squared_error(y_test,prediction)

    rmse=np.sqrt(mse)

    r2=r2_score(y_test,prediction)

    results.append({

        "Model":name,

        "MAE":round(mae,2),

        "MSE":round(mse,2),

        "RMSE":round(rmse,2),

        "R2 Score":round(r2,4)

    })
```

```python
# -------------------- Result --------------------

results_df=pd.DataFrame(results)

results_df=results_df.sort_values(
    by="R2 Score",
    ascending=False
)

print("\n========== MODEL COMPARISON ==========\n")

print(results_df)
```

```python
# -------------------- Best Model --------------------

best=results_df.iloc[0]

print("\n=====================================")
print("Best Regression Model")
print("=====================================")

print("Model :",best["Model"])
print("R2 Score :",best["R2 Score"])
print("RMSE :",best["RMSE"])
print("MAE :",best["MAE"])
```

```python
# -------------------- Save Dataset --------------------

df.to_csv("Synthetic_Regression_Dataset.csv",index=False)

print("\nDataset Saved Successfully!")
```
